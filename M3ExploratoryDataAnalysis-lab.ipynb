## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df['ConvertedComp'].plot(kind='box',figsize=(10,8))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].describe()
```

Find out the upper and lower bounds.


```python
# your code goes here
Q1 = df["ConvertedComp"].quantile(0.25)
Q3 = df["ConvertedComp"].quantile(0.75)
IQR = Q3 - Q1
print(IQR)

```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outlier =[]
for x in df['ConvertedComp']:
  if ((x > Q3) or (x<Q1)):
    outlier.append(x)
print(' outlier in the ConvertedComp column is', len(outlier))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_removed_outlier = df['ConvertedComp'][~((df['ConvertedComp'] <Q1) | 
                                           (df['ConvertedComp'] >Q3))]
```

```python
df_removed_outlier.describe()
```
